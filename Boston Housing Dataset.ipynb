### VisualizaÃ§Ã£o dos dados

```python
# FunÃ§Ã£o para criaÃ§Ã£o de Boxplots

def boxplot(coluna, titulo):
    plt.figure(figsize = (12, 10))
    sns.boxplot(data = coluna, orient = 'h', width = 0.5)
    plt.title(f'{titulo}')
```

```python
# FunÃ§Ã£o para criaÃ§Ã£o de Histogramas

def histograma(coluna, titulo):
    plt.figure(figsize = (12, 10))
    sns.histplot(data = coluna, kde = False, bins = 50)
    plt.title(f'{titulo}')
```

### Algoritmo de Baseline - Random Forest Regressor

Esse algoritmo serÃ¡ ajustado com todas as features do dataset, a fim de comoarÃ¡-lo posteriormente com outro algoritmo composto pelas features de maior importÃ¢ncia.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

```python
X = dataset.drop(['medv'], axis = 1)
y = dataset.medv
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.25, random_state = 424242)
```

```python
regressor = RandomForestRegressor(random_state = 424242)
```

```python
regressor.fit(X_treino, y_treino)
```

```python
print(f'Score de regressÃ£o = {100*regressor.score(X_teste, y_teste):.2f} %')
```

```python
previsoes = regressor.predict(X_teste)
```

```python
print(f'R^2 de regressÃ£o = {100*metrics.r2_score(y_teste, previsoes)} %')
```

```python
print(f'Erro quadrÃ¡tico mÃ©dio = {mean_squared_error(y_teste, previsoes)}')
```

### DeterminaÃ§Ã£o das variÃ¡veis mais importantes utilizando RFECV e o Random Forest como estimador

```python
from sklearn.feature_selection import RFECV
```

```python
# Instanciando o regressor Random Forest

regressor = RandomForestRegressor(random_state = 424242)

# Treinando o modelo

regressor.fit(X_treino, y_treino)

# Instanciando o RFECV e definindo o regressor como mÃ©todo estimador

selecionar_por_rfecv = RFECV(estimator = regressor, cv = 5, scoring = 'r2', step = 1)

# Ajustando o RFECV aos dados de treino e teste

selecionar_por_rfecv.fit(X_treino, y_treino)

# Dados de treino e teste da variÃ¡vel X somente com as variÃ¡veis mais importantes de acordo com o RFECV

X_treino_rfecv = selecionar_por_rfecv.transform(X_treino)

X_teste_rfecv = selecionar_por_rfecv.transform(X_teste)

# Instanciando e ajustando novamente o regressor, dessa vez com os dados de treino X transformados pelo RFECV

regressor = RandomForestRegressor(random_state = 424242)

regressor.fit(X_treino_rfecv, y_treino)

# Score

print(f'Score de regressÃ£o = {100*regressor.score(X_teste_rfecv, y_teste):.2f} %')
```

```python
previsoes = regressor.predict(X_teste_rfecv)
```

```python
print(f'R^2 de regressÃ£o = {100*metrics.r2_score(y_teste, previsoes)} %')
```

```python
print(f'Erro quadrÃ¡tico mÃ©dio = {mean_squared_error(y_teste, previsoes)}')
```

```python
# De acordo com o mÃ©todo RFECV, as colunas mais importantes foram as seguintes

print(f'VariÃ¡veis consideradas mais importantes: {X_treino.columns[selecionar_por_rfecv.support_]}', end = '\n')
print()
print(f'VariÃ¡veis retiradas: {X_treino.columns[~selecionar_por_rfecv.support_]}')
```

### DeterminaÃ§Ã£o das variÃ¡veis mais importantes utilizando o mÃ©todo SelectKBest

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
k_melhores = SelectKBest(f_regression, k = 9)
```

```python
k_melhores.fit(X_treino, y_treino)
```

```python
X_treino_kbest = k_melhores.transform(X_treino)

X_teste_kbest = k_melhores.transform(X_teste)
```

```python
regressor = RandomForestRegressor(random_state = 424242)

regressor.fit(X_treino_kbest, y_treino)

# Score

print(f'Score de regressÃ£o = {100*regressor.score(X_teste_kbest, y_teste):.2f} %')
```

```python
previsoes = regressor.predict(X_teste_kbest)
print(f'R^2 de regressÃ£o = {100*metrics.r2_score(y_teste, previsoes)} %')
```

```python
print(f'VariÃ¡veis consideradas mais importantes: {X_treino.columns[k_melhores.get_support()]}', end = '\n')
print()
print(f'VariÃ¡veis retiradas: {X_treino.columns[~k_melhores.get_support()]}', end = '\n')
```

```python
print(f'Erro quadrÃ¡tico mÃ©dio = {mean_squared_error(y_teste, previsoes)}')
```

# ConclusÃ£o

Foram ajustados trÃªs algoritmos diferentes Ã  base de dados.

O primeiro foi o algoritmo de base: ajustado utilizando todas as features do dataset.
    
O segundo foi o algoritmo ajustado apenas com as variÃ¡veis escolhidas pelo mÃ©todo RFECV.

O terceiro foi o algoritmo ajustado apenas com 9 variÃ¡veis escolhidas pelo mÃ©todo SelectKBest.

Os trÃªs algoritmos apresentaram os parÃ¢metros RÂ² e erros quadrÃ¡ticos mÃ©dios bastante prÃ³ximos uns dos outros, salvo o mÃ©todo SelectKBest, cujo resultado foi um pouco pior. 

Dessa forma, a utilizaÃ§Ã£o entre o algoritmo de base e o algoritmo RFECV fica a critÃ©rio do usuÃ¡rio, a depender das informaÃ§Ãµes que lhe estiverem disponÃ­veis.
