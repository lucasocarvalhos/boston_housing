# Boston Housing Dataset

## Objetivo: Predizer a mediana do valor de casas ocupadas por moradores.

Dataset: https://www.cs.toronto.edu/~delve/data/boston/bostonDetail.html

```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('Boston.csv')
```

```python
dataset.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

### Visualização dos dados

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Função para criação de Boxplots

def boxplot(coluna, titulo):
    plt.figure(figsize = (12, 10))
    sns.boxplot(data = coluna, orient = 'h', width = 0.5)
    plt.title(f'{titulo}')
```

```python
# Função para criação de Histogramas

def histograma(coluna, titulo):
    plt.figure(figsize = (12, 10))
    sns.histplot(data = coluna, kde = False, bins = 50)
    plt.title(f'{titulo}')
```

```python
for item in dataset.columns:
    boxplot(dataset[item], f'Distribuição da variável {item}')
```

```python
for item in dataset.columns:
    histograma(dataset[item], f'Frequências da variável {item}')
```

```python
# Matriz de correlação

plt.figure(figsize = (12, 10))
sns.heatmap(dataset.corr(), annot = True, fmt = '.1f')
```

```python
# Pairplot de todas as variáveis com a variável medv

sns.pairplot(y_vars = 'medv', x_vars = dataset.columns, data = dataset)
```

### Algoritmo de Baseline - Random Forest Regressor

Esse algoritmo será ajustado com todas as features do dataset, a fim de comoará-lo posteriormente com outro algoritmo composto pelas features de maior importância.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

```python
X = dataset.drop(['medv'], axis = 1)
y = dataset.medv
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.25, random_state = 424242)
```

```python
regressor = RandomForestRegressor(random_state = 424242)
```

```python
regressor.fit(X_treino, y_treino)
```

```python
print(f'Score de regressão = {100*regressor.score(X_teste, y_teste):.2f} %')
```

```python
previsoes = regressor.predict(X_teste)
```

```python
print(f'R^2 de regressão = {100*metrics.r2_score(y_teste, previsoes)} %')
```

```python
print(f'Erro quadrático médio = {mean_squared_error(y_teste, previsoes)}')
```

### Determinação das variáveis mais importantes utilizando RFECV e o Random Forest como estimador

```python
from sklearn.feature_selection import RFECV
```

```python
# Instanciando o regressor Random Forest

regressor = RandomForestRegressor(random_state = 424242)

# Treinando o modelo

regressor.fit(X_treino, y_treino)

# Instanciando o RFECV e definindo o regressor como método estimador

selecionar_por_rfecv = RFECV(estimator = regressor, cv = 5, scoring = 'r2', step = 1)

# Ajustando o RFECV aos dados de treino e teste

selecionar_por_rfecv.fit(X_treino, y_treino)

# Dados de treino e teste da variável X somente com as variáveis mais importantes de acordo com o RFECV

X_treino_rfecv = selecionar_por_rfecv.transform(X_treino)

X_teste_rfecv = selecionar_por_rfecv.transform(X_teste)

# Instanciando e ajustando novamente o regressor, dessa vez com os dados de treino X transformados pelo RFECV

regressor = RandomForestRegressor(random_state = 424242)

regressor.fit(X_treino_rfecv, y_treino)

# Score

print(f'Score de regressão = {100*regressor.score(X_teste_rfecv, y_teste):.2f} %')
```

```python
previsoes = regressor.predict(X_teste_rfecv)
```

```python
print(f'R^2 de regressão = {100*metrics.r2_score(y_teste, previsoes)} %')
```

```python
print(f'Erro quadrático médio = {mean_squared_error(y_teste, previsoes)}')
```

```python
# De acordo com o método RFECV, as colunas mais importantes foram as seguintes

print(f'Variáveis consideradas mais importantes: {X_treino.columns[selecionar_por_rfecv.support_]}', end = '\n')
print()
print(f'Variáveis retiradas: {X_treino.columns[~selecionar_por_rfecv.support_]}')
```

### Determinação das variáveis mais importantes utilizando o método SelectKBest

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
k_melhores = SelectKBest(f_regression, k = 9)
```

```python
k_melhores.fit(X_treino, y_treino)
```

```python
X_treino_kbest = k_melhores.transform(X_treino)

X_teste_kbest = k_melhores.transform(X_teste)
```

```python
regressor = RandomForestRegressor(random_state = 424242)

regressor.fit(X_treino_kbest, y_treino)

# Score

print(f'Score de regressão = {100*regressor.score(X_teste_kbest, y_teste):.2f} %')
```

```python
previsoes = regressor.predict(X_teste_kbest)
print(f'R^2 de regressão = {100*metrics.r2_score(y_teste, previsoes)} %')
```

```python
print(f'Variáveis consideradas mais importantes: {X_treino.columns[k_melhores.get_support()]}', end = '\n')
print()
print(f'Variáveis retiradas: {X_treino.columns[~k_melhores.get_support()]}', end = '\n')
```

```python
print(f'Erro quadrático médio = {mean_squared_error(y_teste, previsoes)}')
```

# Conclusão

Foram ajustados três algoritmos diferentes à base de dados.

O primeiro foi o algoritmo de base: ajustado utilizando todas as features do dataset.
    
O segundo foi o algoritmo ajustado apenas com as variáveis escolhidas pelo método RFECV.

O terceiro foi o algoritmo ajustado apenas com 9 variáveis escolhidas pelo método SelectKBest.

Os três algoritmos apresentaram os parâmetros R² e erros quadráticos médios bastante próximos uns dos outros, salvo o método SelectKBest, cujo resultado foi um pouco pior. 

Dessa forma, a utilização entre o algoritmo de base e o algoritmo RFECV fica a critério do usuário, a depender das informações que lhe estiverem disponíveis.
